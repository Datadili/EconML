```python
import numpy as np
def activition_function(vector, activation_func:str):
    if activation_func == "ReLU":
        #loops through elements in vector to see the maximum btw the element and 0
        g_x = np.maximum(vector, 0)
        
    elif activation_func == "sigmoid":
        g_x = 1/(1 + np.exp(-vector))
    return g_x
```

```python
import matplotlib.pyplot as plt

vector = np.linspace(-5, 5, 300)

#np.random.randint(-10, 11, size=n)
act_funcs = ["sigmoid", "ReLU"]
for act_function in act_funcs:
    function_values = activition_function(vector, act_function)
    plt.plot(vector, function_values)

plt.ylim(0, 1)
plt.show()
```

```python
#implement Fibonacci1 and Fibonacci2
import numpy as np
def Fibonacci1(n):
    n = int(n)
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return Fibonacci1(n-1)+Fibonacci1(n - 2)
    
def Fibonacci2(n):
    n = int(n)
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        v = np.zeros(n+1)
        v[0] = 0
        v[1] = 1
        for i in range(2, n+1):
            v[i] = v[i-1] + v[i - 2]
        return v[n]
```

```python
import time
N = np.linspace(1,40, 40)

Fibonacci1s = []
time_Fibonacci1s = []
Fibonacci2s = []
time_Fibonacci2s = []

for n in N:
    start_time = time.time() 
    Fibonacci1(n)
    stop_time = time.time() 
    time_Fibonacci1s.append(stop_time - start_time)
    
for n in N:
    start_time = time.time() 
    Fibonacci2(n)
    stop_time = time.time() 
    time_Fibonacci2s.append(stop_time - start_time)  
```

```python
import matplotlib.pyplot as plt
import math

# Assuming Fibonacci1s, time_Fibonacci1s, Fibonacci2s, time_Fibonacci2s are defined

# Create subplots
fig, axs = plt.subplots(2)

# Plot for Fibonacci1s
axs[0].plot(N, time_Fibonacci1s)
axs[0].set_title('Fibonacci 1')
axs[0].set_xlabel('Fibonacci Number')
axs[0].set_ylabel('Execution Time (seconds)')

# Plot for Fibonacci2s
axs[1].plot(N, time_Fibonacci2s)
axs[1].set_title('Fibonacci 2')
axs[1].set_xlabel('Fibonacci Number')
axs[1].set_ylabel('Execution Time (seconds)')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

**Ans:** Fibonacci 2 is faster
